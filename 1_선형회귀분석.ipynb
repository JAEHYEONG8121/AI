## 선형회귀분석
> 선형회귀분석을 간단히 설명하자면, 주어진 데이터를 가장 잘 설명하는 직선 하나를 찾는 것이라고 할 수 있다.  

> 독립변수의 수에 따라서 한 개인 경우 '단순선형회귀', 여러 개인 경우 '다중선형회귀'로 나눌 수 있다.  

> 밑의 그림과 같이 단순선형회귀분석을 한다는 것은 <font color=blue>x와 y라는 데이터가 주어졌을 때, y = w * x + b라는 직선의 방정식에서 데이터를 가장 잘 표현하는 변수 w와 b를 찾는다</font>는 뜻이다!  

> 이 때 w와 b는 각각 가중치(weight)와 편차(bias)를 각각 한 문자로 표현한 것이다. 수학 용어로 설명하면 '가중치 = 기울기, 편차 = 절편'인 셈이다.

![선형회귀분석](img/LR.jpg)
***

***손실 함수(Loss function) 및 경사하강법(Gradient descent)***
> 데이터를 '가장 잘' 표현하는 직선을 찾는다고 했는데, '가장 잘' 표현한다는 말에는 비교가 가능하다는 뜻이 깔려있다. 즉 어떤 w, b쌍에 대하여 데이터와 얼마나 잘 맞는지 수치적으로 계산을 할 수 있어야 한다.  

****평균제곱오차(Mean Squared Error)***

![MSE](img/2.png)

- 수식의 의미는 예측한 값과 셀제 데이터 값의 차이를 제곱하여 평균을 낸다는 것이다.
- 예측한 값은 w와 b에 의해 정해지기 때문에 w, b쌍들을 비교하는 것이 가능해진다.
- 이제 비교가 가능해졌기 때문에 이제 목표는 지정한 척도인 MSE를 최소화하는 w, b쌍을 찾는 것이다!  

****손실 함수(Loss function) or 비용 함수(Cost function)***
> 그렇다면, 어떻게 최적의 w, b쌍을 찾을 수 있을까?  

> 오차는 예측값과 목푯값의 차이이며, 이를 나타내는 것이 바로 '손실 함수'이다. 여기서 예측값은 w와 b에 의해 구해지므로 이러한 관계를 통해 최적의 w, b를 찾는 방법을 생각해볼 수 있다.  

![L2](img/1.png)

- 위의 수식은 MSE를 손실 함수로 사용했을 때(이를 L2 손실함수라고 부른다) 변수 w에 대한 오차를 나타낸 것이다.
- 결국, 위 수식은 변수 w에 대한 2차 함수이다. 이런 상황에서 오차를 최소화하는 w를 어떻게 찾을까?  

****경사하강법(Gradient desccent)***
> w에 대한 2차 함수 형태이므로 딱 보기에는 w로 미분해서 그 값이 0인 지점을 찾으면 되지 않을까라는 생각을 해볼 수 있다. 하지만, w를 구하려면 연산의 복잡도가 너무 커진다.  

> 따라서, 대부분 딥러닝 모델에서는 '경사하강법'을 이용한다
여기서 경사는 '기울기'로서 미분 값을 의미하고, 미분 값이라는 것은 2차 함수의 어느 한 지점에서의 순간적인 기울기를 의미한다.  

> 즉, 주어진 w에서 경사를 구하고 이를 통해 지속적으로 w를 업데이트함으로써 오차의 극솟값을 찾을 수 있는데 이를 수식으로 표현하면 아래와 같다.

![경사하강법](img/3.png)

- 위 식의 논리가 무엇일까. 특정 포인트 w에서 w가 커질 수록 함수값이 커지는 중이라면(즉, 기울기 양수) 음의 방향으로 w를 옮겨야 할 것이고, 반대로 특정 포인트 w에서 w가 커질 수록 함수값이 작아지는 중이라면(즉, 기울기 음수) 양의 방향으로 w를 옮기면 된다.
- 그렇다면, 'learning rate = step size = 이동거리'는 어떻게 구할 것인가? 
- 이동거리는 gradient값을 구하는 지점을 얼만큼 옮길 지에 대한 척도인데, 잘 생각해보면 미분 계수(gradient값)값은 극솟값에 가까울 수록 작다.
- 그러므로, 현재의 w값이 극솟값에서 멀 때는 많이 이동해야하고, 극솟값에 가까워졌을 때는 조금씩 이동할 수 있게 된다.
- 따라서, 이동거리는 gradient값을 직접 이용하되, 이동거리를 적절히 사용자가 조절 할 수 있게 수식을 조정해 줌으로써 상황에 맞게 이동거리를 맞춰나갈 수 있게 하면 된다!  

****이동거리(step size)에 대한 고찰***
> step size가 큰 경우 한 번 이동하는 거리가 커지므로 빠르게 수렴할 수 있다는 장점이 있다. 하지만, step size를 너무 크게 설정해버리면 최소값을 계산하도록 수렴하지 못하고 함수 값이 계속 커지는 방향으로 발산해버릴 수 있다.  
> 반대로 step size가 너무 작은 경우에는 매 step별로 이동하는 거리가 너무 작아서 수렴하지 못하게 된다.

> 텐서는 n차원의 배열을 전부 포함하는 넓은 개념이다.

```python
# 텐서 생성
import torch

X = torch.Tensor(2, 3)
X = torch.tensor([[1, 2, 3], [4, 5, 6]])
X
```

- 위의 코드를 실행시키면 1행에서는 PyTorch프레임워크를 불러오고 2행에서는 X라는 변수에 파이토치 텐서를 하나 생성해서 지정한다.
- 이때 지정된 텐서의 shape는 2x3인 것이다.
- 텐서를 생성하면서 원하는 값으로 초기화하려면 인수로 배열을 전달하면된다.

***
****torch.tensor()함수의 인수***
> torch.tensor 함수는 인수로 data, dytpe, device, requires_grad등을 받는다.
1. data : 앞서 본 것처럼 배열이 들어간다.
2. dtype : 데이터를 저장할 자료형이 들어간다. (기본값 : FloatTensor)
3. device : 이 텐서를 어느 기기에 올릴 것인지를 명시한다.
4. requires_grad : 이 텐서에 대한 기울기를 저장할지 여부를 지정한다. (기본값 : False)

```python
import torch

x_tensor = torch.tensor(data=[2.0, 3.0], requires_grad=True)
x_tensor
```

***
****기울기 계산 코드***
> 위에서 생성한 텐서를 가지고 연산 그래프를 생성하면 연산 그래프는 어떠한 결과를 산출할 것이다.   

> 선형회귀분석 모델을 생성하고 경사하강법을 적용한 전체 과정 코드를 보기 전에 우선 기울기를 계산하는 코드를 살펴보자.

```python
# z = 2x^2 + 3 식에서 x에 대한 기울기를 구하기
import torch

x = torch.tensor(data=[2.0, 3.0], requires_grad=True)
y = x**2
z = 2*y + 3

target = torch.tensor([3.0, 4.0])
loss = torch.sum(torch.abs(z - target))
loss.backward()

print(x.grad, y.grad, z.grad)
```

- x라는 텐서를 생성하여 기울기를 계산하도록 지정하였다.
- z라는 변수에는 연산 그래프의 결과값이 저장된다.
- z와 목표값이 target의 절댓값 차이를 계산하고 'torch.sum()'이란 함수를 통해 두 값이 차이를 숫자 하나로 바꾼다 -> L1 Loss를 구한 것이다!
- 그 다음 'loss.backward()'함수를 호출하면 연산 그래프를 쭉 따라가면서 잎 노드(leaf node) x에 대한 기울기를 계산한다.
- 여기서 잎 노드는 다른 변수를 통해 계산되는 y나 z가 아니라, 그 자체가 값인 x 같은 노드를 의미한다.
- 결과적으로 x.grad는 기울기가 계산되지만 y.grad, z.grad는 잎 노드가 아니기 때문에 결과값이 None이 리턴된다.
***

****참고) L1 Loss & L2 Loss***
> L1 Loss : 두 개의 벡터를 빼고, 절댓값을 취한 뒤 합한 것이다.

![L1 Loss](img/L1.png)

> L2 Loss : 두 개 벡터(점) 사이의 직선 거리를 의미한다.

![L2 Loss](img/L2.png)

- L1 Loss는 L2 Loss에 비해 이상치의 영향을 덜 받는 Robust한 특성을 가지지만, 0에서 미분이 불가능하다.
- L2 Loss는 L1 Loss와 달리 제곱을 취하므로, 이상치가 들어오면 오차가 제곱이 되어 이상치에 더 영향을 받는다.

***
****선형회귀분석 모델을 생성 -> 기울기 계산 -> w, b를 업데이트하는 전체 과정 코드***
- 선형 관계를 가지고 있는 데이터(Linear Data)
- 선형 모델(Linear Model)
- y = 2x + 3

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import numpy as np
import matplotlib.pyplot as plt
```

- 1. torch.nn : 신경망 모델들이 포함되어 있다.
- 2. torch.optim : 경사하강법 알고리즘이 포함되어 있다.
- 3. torch.nn.init : 텐서에 초깃값을 주기 위한 함수들이 포함되어 있다.

```python
num_data = 1000 # 데이터 수
num_epoch = 500 # 경사하강법 반복 횟수

x = init.uniform_(torch.Tensor(num_data, 1), -10, 10)
noise = init.normal(torch.FloatTensor(num_data, 1), std=1)
y = 2*x + 3
y_noise = y + noise
```

- 사용할 데이터 수를 num_data에, 경사하강법 반복 횟수를 num_epoch변수에 각각 저장한다.
- x라는 변수에 [num_data, 1] shpae의 텐서를 생성한다.
- 이 텐서의 값들을 'init.uniform_()'함수를 통해 -10 ~ 10사이의 값을 균등(uniform)하게(무작위로) 초기화한다.
- 그러므로, y는 -17 ~ 23사이의 값으로 분포하게 된다.  
 
- y에 노이즈를 추가하기 위해 y_noise변수를 생성한다.
    - <노이즈를 추가하는 이유> : 데이터에 어떤 관계가 존재한다고 해도 보통 센서나 관측을 통해 들어오는 데이터에는 노이즈가 추가된 상태로 들어오는 경우 대부분이기 때문에, 현실성을 반영하기 위해 추가한 것이다! 
         
- 이때 노이즈는 표준정규분포를 따르는 '가우시안 노이즈'를 사용한다
- 또한, y와 같은 shape를 갖는 노이즈 텐서를 만들어야 하므로 [num_data, 1] shape 텐서 noise를 'init.normal_()'함수를 통해 초기화해준다. -> 이떄 mean(평균) = 0(default값), std(표준편차) = 1로 지정하였고, y에다가 이 noise를 더해서 만든 갑이 y_noise변수에 저장한다.

```python
model = nn.Linear(1, 1)
loss_func = nn.L1Loss()
```

- 데이터가 준비되었으니 선형회귀모델을 만들자. 파이토치에서 선형회귀 모델은 'nn.Linear()'라는 함수로 구현되어 있다.
- Linear 클래스의 인수 & 변수  

   > <인수>
   > - 1.들어오는 특성(feature)의 수
   > - 2.결과로 나오는 특성의 수
   > - 3.편차 사용 여부 <br/><br/>
     
   
   > <변수>
   > - 가중치(weight), 편차(bias) <br/><br/>
  

- 우리가 만든 데이터 x는 1개의 특성을 갖는 데이터 1000개이고, 결과 y도 1개의 특성을 갖는 데이터 1000개이므로 인수로 모델을 Linear(1, 1)로 생성한다.
- 모델에서 나온 결과와 y_noise와의 차이를 구하는 척도로는 차이의 절댓값의 평균인 L1 Loss를 사용했다.
            

```python
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

- 경사하강법을 사용하기 위해서 torch.optim에서 SGD 옵티마이저를 불러온다.
- 옵티마이저(optimizer) = 최적화 함수(optimization function)
- 최적화할 변수들(w, b)와 함께, 학습률 lr(learning rate)를 인수로 전달한다.
- 최적화할 변수는 'model.parameters()'라는 함수를 이용하여 변수를 전달했고, 학습률로는 0.01이라는 수치를 전달했다.

```python
loss_arr = [] # 손실이 어떻게 변하는지 확인하기 위해 리스트 생성

label = y_noise

for i in range(num_epoch) :
    optimizer.zero_grad()
    output = model(x)
    
    loss = loss_func(output, label)
    loss.backward()
    optimizer.step()
    
    if i % 10 == 0 :
        print(loss.data)
    param_list = list(model.parameters())
    print("Weight:",param_list[0].item(), "\nBias:  ", param_list[1].item(), '\n')
    
    loss_arr.append(loss.detach().numpy())
```

- 1. num_epoch
    - 경사하강법을 반복하는 횟수이다.
    - 에폭(epoch) : 데이터 전체를 학습에 한 번 사용하는 주기</br></br>
- 2. 기울기 초기화
    - optimizer.zero_grad() 함수를 통해 반복문에서 반복 시 지난번에 계산했던 기울기를 0으로 초기화해준다.</br></br>
- 3. 선형회귀 모델
    - 기울기를 초기화해야 새로운 가중치와 편차에 대한 새로운 기울기를 구할 수 있기 때문이다.
    - 선형회귀 모델에 x를 전달하고 이를 output변수에 담는다.</br></br>
- 4. L1 Loss를 통한 Loss 계산
    - L1 손실 함수의 정의에 따라 output과 y_noise의 차이를 loss에 저장한다.</br></br>
- 5. Loss에 대한 미분 연산
    - loss.backward()함수를 통해서 w, b에 대한 기울기가 계산된다.</br></br>
- 6. 경사하강법 최적화
    - optimizer의 step()함수를 통해 인수로 들어갔던model.parameters()에서 리턴되는 변수들의 기울기에 학습률 0.01을 곱하여 빼줌으로써 업데이트한다.

```python
plt.figure(figsize=(15, 15))
plt.scatter(x.numpy(), y_noise.numpy(), s=5, c="gray")
plt.scatter(x.detach().numpy(), output.detach().numpy(), s=5, c="red")
plt.axis([-10, 10, -30, 30])
plt.show()
```
